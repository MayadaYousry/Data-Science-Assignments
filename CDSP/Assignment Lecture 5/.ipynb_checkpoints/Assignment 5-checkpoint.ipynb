```python
import requests

def get_current_temp(lat,long):
    
    api_key = "bea34a996c8eb297edf4f2a22fe5bb91"
    lat = str(lat)
    lon = str(long)
    url = "https://api.openweathermap.org/data/2.5/onecall?lat=%s&lon=%s&appid=%s&units=metric" % (lat, lon, api_key)

    response = requests.get(url)
    data = response.json()
    return data['current']['temp']

df['Temp_Diff'] = df['Temperature'] - df.apply(lambda x: get_current_temp(x['Destination Lat'], x['Destination Long']), axis=1)
df
```

# Task 2

```python
df = pd.read_csv("ufo.csv")
df = df.head(100)
df
```

```python
df['Time'] = pd.to_datetime(df["Time"])
```

```python
df['IN_December'] = (df['Time'].dt.month).apply(lambda x: x == 12)
df
```

```python
import datetime
from dateutil.relativedelta import relativedelta

WW1 = datetime.datetime(1914, 7, 28)
df['Time_Diff_From_WW1'] = df['Time'].apply(lambda x: relativedelta(x, WW1).years) 
df
```

# Task3

```python
df = pd.read_csv("visitor-interests.csv")
df = df.head(50)
df
```

```python
import user_agents

def extract_web_data(ua):
    ua = user_agents.parse(ua)
    return pd.Series((ua.browser.family, ua.os.family, ua.device.family))
df[['Browser', 'OS', 'Device']] = df['UserAgent'].apply(extract_web_data)
df
```

```python
import requests
import json

def get_lat_long(ip):
    response = requests.get("https://geolocation-db.com/jsonp/" + ip)
    result = response.content.decode()
    result = result.split("(")[1].strip(")")
    result = json.loads(result)
    return pd.Series((result['latitude'], result['longitude']))

df[['Latitude', 'Longitude']] = df['IP'].apply(get_lat_long)
df
```
