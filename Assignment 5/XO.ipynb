```python
import random

class XO_Game:
    '''
        mode 1: human to human
        mode 2: human to computer
        mode 3: human to computer intelligent
        mode 4 computer to computer intelligent
    ''' 
    def __init__(self, mode = 1):
        if(mode in [1,2,3,4]):
            self.mode = mode
            self.__structure = ["1", "2", "3", "4", "5", "6", "7", "8", "9"]
            self.__busy_cells = []
            self.__is_game_on = True
            self.__current_player = "X"
            self.__play_game()
        else:
            print("Invalid mode number!")
        
        
    def __draw_structure(self):
      print("\n")
      print(self.__structure[0] + " | " + self.__structure[1] + " | " + self.__structure[2])
      print("----------")
      print(self.__structure[3] + " | " + self.__structure[4] + " | " + self.__structure[5])
      print("----------")
      print(self.__structure[6] + " | " + self.__structure[7] + " | " + self.__structure[8])
      print("\n")
    
    def __toggle_player(self):
        if self.__current_player == "X":
            self.__current_player = "O"
            if self.mode == 2:
                self.__play_computer()
            elif self.mode == 3:
                self.__play_computer_intelligent()
        else:
            self.__current_player = "X"
            
    def __play_game(self):
        self.__draw_structure()
        while self.__is_game_on:
            if self.mode == 4:
                self.__play_computer_intelligent()
            else:
                correct_value = False
                while correct_value == False:
                    print(f"{self.__current_player}'s turn\n")
                    try:
                        position = int(input("Please enter position\n"))
                        if(position < 1 or position > 9):
                            print("Please enter value between 1 and 9")
                            correct_value = False
                        elif self.__busy_cells.count(position) > 0:
                            print("Please enter value of empty position\n")
                            correct_value = False
                        else:
                            self.__busy_cells.append(position)
                            self.__structure[position - 1] = self.__current_player
                            self.__draw_structure()
                            self.__get_winner()
                            if self.__is_game_on:
                                self.__toggle_player()
                            correct_value = True
                    except ValueError:
                        print("Please enter correct value\n")
                        correct_value = False
                    
            
    def __play_computer(self):
        print(f"{self.__current_player}'s turn\n")

        empty_cells = list({1,2,3,4,5,6,7,8,9}-set(self.__busy_cells))
        position = random.choice(empty_cells)

        self.__busy_cells.append(position)
        self.__structure[position - 1] = self.__current_player
        self.__draw_structure()
        self.__get_winner()
        self.__toggle_player()
        
    def __play_computer_intelligent(self):
        print(f"{self.__current_player}'s turn\n")

        empty_cells = list({1,2,3,4,5,6,7,8,9}-set(self.__busy_cells))
        position = -1

        for cell in empty_cells:
            #check for positions that make computer win
            if cell == 1:
                if self.__structure[1] == self.__structure[2] == self.__current_player or self.__structure[3] == self.__structure[6] == self.__current_player or self.__structure[4] == self.__structure[8] == self.__current_player:
                    position = 1
                    break
            elif cell == 2:
                if self.__structure[0] == self.__structure[2] == self.__current_player or self.__structure[4] == self.__structure[7] == self.__current_player:
                    position = 2
                    break
            elif cell == 3:
                if self.__structure[0] == self.__structure[1] == self.__current_player or self.__structure[5] == self.__structure[8] == self.__current_player or self.__structure[4] == self.__structure[6] == self.__current_player:
                    position = 3
                    break
            elif cell == 4:
                if self.__structure[0] == self.__structure[6] == self.__current_player or self.__structure[4] == self.__structure[5] == self.__current_player:
                    position = 4
                    break
            elif cell == 5:
                if self.__structure[1] == self.__structure[7] == self.__current_player or self.__structure[3] == self.__structure[5] == self.__current_player or self.__structure[0] == self.__structure[8] == self.__current_player or self.__structure[2] == self.__structure[6] == self.__current_player:
                    position = 5
                    break
            elif cell == 6:
                if self.__structure[2] == self.__structure[8] == self.__current_player or self.__structure[3] == self.__structure[4] == self.__current_player:
                    position = 6
                    break
            elif cell == 7:
                if self.__structure[0] == self.__structure[3] == self.__current_player or self.__structure[7] == self.__structure[8] == self.__current_player or self.__structure[2] == self.__structure[4] == self.__current_player:
                    position = 7
                    break
            elif cell == 8:
                if self.__structure[1] == self.__structure[4] == self.__current_player or self.__structure[6] == self.__structure[8] == self.__current_player:
                    position = 8
                    break
            elif cell == 9:
                if self.__structure[2] == self.__structure[5] == self.__current_player or self.__structure[6] == self.__structure[7] == self.__current_player or self.__structure[0] == self.__structure[4] == self.__current_player:
                    position = 9
                    break

            #check for positions that make the human player win
            if cell == 1:
                if self.__structure[1] == self.__structure[2] != self.__current_player or self.__structure[3] == self.__structure[6] != self.__current_player or self.__structure[4] == self.__structure[8] != self.__current_player:
                    position = 1
                    break
            elif cell == 2:
                if self.__structure[0] == self.__structure[2] != self.__current_player or self.__structure[4] == self.__structure[7] != self.__current_player:
                    position = 2
                    break
            elif cell == 3:
                if self.__structure[0] == self.__structure[1] != self.__current_player or self.__structure[5] == self.__structure[8] != self.__current_player or self.__structure[4] == self.__structure[6] != self.__current_player:
                    position = 3
                    break
            elif cell == 4:
                if self.__structure[0] == self.__structure[6] != self.__current_player or self.__structure[4] == self.__structure[5] != self.__current_player:
                    position = 4
                    break
            elif cell == 5:
                if self.__structure[1] == self.__structure[7] != self.__current_player or self.__structure[3] == self.__structure[5] != self.__current_player or self.__structure[0] == self.__structure[8] != self.__current_player or self.__structure[2] == self.__structure[6] != self.__current_player:
                    position = 5
                    break
            elif cell == 6:
                if self.__structure[2] == self.__structure[8] != self.__current_player or self.__structure[3] == self.__structure[4] != self.__current_player:
                    position = 6
                    break
            elif cell == 7:
                if self.__structure[0] == self.__structure[3] != self.__current_player or self.__structure[7] == self.__structure[8] != self.__current_player or self.__structure[2] == self.__structure[4] != self.__current_player:
                    position = 7
                    break
            elif cell == 8:
                if self.__structure[1] == self.__structure[4] != self.__current_player or self.__structure[6] == self.__structure[8] != self.__current_player:
                    position = 8
                    break
            elif cell == 9:
                if self.__structure[2] == self.__structure[5] != self.__current_player or self.__structure[6] == self.__structure[7] != self.__current_player or self.__structure[0] == self.__structure[4] != self.__current_player:
                    position = 9
                    break

        if position < 0:
            position = random.choice(empty_cells)

        self.__busy_cells.append(position)
        self.__structure[position - 1] = self.__current_player
        self.__draw_structure()
        self.__get_winner()
        if self.__is_game_on:
            self.__toggle_player()
                    
    def __get_winner(self):
        if(self.__structure[0] == self.__structure[1] == self.__structure[2] or
           self.__structure[3] == self.__structure[4] == self.__structure[5] or
           self.__structure[6] == self.__structure[7] == self.__structure[8] or
           self.__structure[0] == self.__structure[3] == self.__structure[6] or 
           self.__structure[1] == self.__structure[4] == self.__structure[7] or
           self.__structure[2] == self.__structure[5] == self.__structure[8] or
           self.__structure[0] == self.__structure[4] == self.__structure[8] or
           self.__structure[2] == self.__structure[4] == self.__structure[6]):
            print(f"{self.__current_player} is the winner!")
            self.__end_game()
        elif(len(self.__busy_cells) == 9):
            print("Game over!\n")
            self.__end_game()
    
    def __end_game(self):
        self.__is_game_on = False
        play_again = input("Play again? Y|N press any key to exit")
        if(play_again.lower() == "y"):
            mode = int(input("Please enter mode number "))
            self.__init__(mode)
```
